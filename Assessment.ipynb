```python
# Add Happiness Category column
def categorize_happiness(score):
    if score < 4:
        return 'Low'
    elif 4 <= score <= 6:
        return 'Medium'
    else:
        return 'High'

data['Happiness Category'] = data['score'].apply(categorize_happiness)
data[['Country name', 'score', 'Happiness Category']].head(10)
```

```python
# Top 10 unhappiest countries
top_10_unhappiest = data.nsmallest(10, 'score')

# Line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Country name', y='score', data=top_10_unhappiest, color='red', marker = "o")
plt.title('Top 10 Unhappiest Countries by Score')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.xticks(rotation=30)
plt.show()
```

```python
# Calculating Q1, Q3, and IQR for the column Score
score_Q1 = df_sa_sorted['score'].quantile(0.25)
score_Q3 = df_sa_sorted['score'].quantile(0.75)
score_IQR = score_Q3 - score_Q1

# Calculating lower and upper bounds for Score outliers
score_lower_bound = score_Q1 - 1.5 * score_IQR
score_upper_bound = score_Q3 + 1.5 * score_IQR

# Calculating Q1, Q3, and IQR for the column GDP per capita
gdp_Q1 = df_sa_sorted['Log GDP per capita'].quantile(0.25)
gdp_Q3 = df_sa_sorted['Log GDP per capita'].quantile(0.75)
gdp_IQR = gdp_Q3 - gdp_Q1

# Calculating lower and upper bounds for GDP per capita outliers
gdp_lower_bound = gdp_Q1 - 1.5 * gdp_IQR
gdp_upper_bound = gdp_Q3 + 1.5 * gdp_IQR

# Identifying outliers
outliers = df_sa_sorted[
    (df_sa_sorted['score'] < score_lower_bound) |
    (df_sa_sorted['score'] > score_upper_bound) |
    (df_sa_sorted['Log GDP per capita'] < gdp_lower_bound) |
    (df_sa_sorted['Log GDP per capita'] > gdp_upper_bound)
    ]
```

```python
# Scatter plot for GDP per Capita vs. Score
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Log GDP per capita', y='score', data=outliers, color='red', label='outliers')

# Customize plot
plt.title('Log GDP per capita vs. Happiness Score (Outliers Highlighted)', fontsize=14)
plt.xlabel('Log GDP per capita', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)
plt.legend(title='Category')
plt.grid(True)
plt.show()
```

1. Descriptive Statistics:

 Which region has higher happiness Scores on average?

 Compare key metrics like GDP per Capita, Social Support, and Healthy Life Expectancy

4. Happiness Disparity:

 Compute the range (max - min) and coefficient of variation (CV) for Score in both regions.

 Which region has greater variability in happiness?

```python
# Range and CV for South Asia
south_asia_range = df_sa_sorted['score'].max() - df_sa_sorted['score'].min()
south_asia_cv = df_sa_sorted['score'].std() / df_sa_sorted['score'].mean()

# Range and CV for Middle East
middle_east_range = middle_east_data['score'].max() - middle_east_data['score'].min()
middle_east_cv = middle_east_data['score'].std() / middle_east_data['score'].mean()

print(f"South Asia - Range: {south_asia_range:.3f}, CV: {south_asia_cv:.3f}")
print(f"Middle East - Range: {middle_east_range:.3f}, CV: {middle_east_cv:.3f}")

# Determine which region has greater variability
if south_asia_cv > middle_east_cv:
    print("South Asia has greater variability in happiness.")
else:
    print("Middle East has greater variability in happiness.")
```

 Create scatter plots to visualize and interpret the relationships.


```python
# Correlation analysis for both regions
for metric in ['Freedom to make life choices', 'Generosity']:
    south_corr = df_sa_sorted['score'].corr(df_sa_sorted[metric])
    middle_corr = middle_east_data['score'].corr(middle_east_data[metric])
    print(f"{metric} - South Asia Correlation: {south_corr:.3f}, Middle East Correlation: {middle_corr:.3f}")

    # Scatter plots
    plt.figure(figsize=(10, 6))
    sns.regplot(x=metric, y='score', data=df_sa_sorted, label='South Asia', color='blue')
    sns.regplot(x=metric, y='score', data=middle_east_data, label='Middle East', color='orange')
    plt.title(f'{metric} vs. Score')
    plt.xlabel(metric)
    plt.ylabel('Happiness Score')
    plt.legend()
    plt.show()
```

6. Outlier Detection:

 Identify outlier countries in both regions based on Score and GDP per Capita.

 Plot these outliers and discuss their implications.

```python
def outliers_iqr(data, column_name):
    q1 = data[column_name].quantile(0.25)
    q3 = data[column_name].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    outliers = data[(data[column_name] < lower_bound) | (data[column_name] > upper_bound)]
    return outliers

me_outlier_score = outliers_iqr(middle_east_data, 'score')
sa_outlier_score = outliers_iqr(df_sa_sorted, 'score')
me_outlier_gdp = outliers_iqr(middle_east_data, 'Log GDP per capita')
sa_outlier_gdp = outliers_iqr(df_sa_sorted, 'Log GDP per capita')
```

```python
sa_concat_oulier = pd.concat([sa_outlier_score, sa_outlier_gdp]).drop_duplicates()
me_concat_oulier = pd.concat([me_outlier_score, me_outlier_gdp]).drop_duplicates()
sa_concat_oulier
```

```python
# Scatter plot for South Asia
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_sa_sorted, x='Log GDP per capita', y='score', label='Non-Outliers', color='blue')
sns.scatterplot(data=sa_concat_oulier, x='Log GDP per capita', y='score', label='Outliers', color='red')
plt.title('South Asia: Outliers Based on Score and GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Scatter plot for Middle East
plt.figure(figsize=(10, 6))
sns.scatterplot(data=middle_east_data, x='Log GDP per capita', y='score', label='Non-Outliers', color='green')
sns.scatterplot(data=me_concat_oulier, x='Log GDP per capita', y='score', label='Outliers', color='orange')
plt.title('Middle East: Outliers Based on Score and GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.legend()
plt.grid(True)
plt.show()
```
